```python
## import the essential module
import pymongo
import logging
import csv
logging.basicConfig(filename='Carbon_Nano.log',level=logging.INFO,format='%(asctime)s %(levelname)s %(message)s')
```

```python
## modular coding
class MongoDB:
    def __init__(self,url):
        self.url=url
        try:
            self.client=pymongo.MongoClient(self.url)
        except Exception as e:
            print('Error has occured' + str(e))
            logging.exception('Error occured' + str(e))
        else:
            logging.info('Connection created' + str(self.client))
            print('Connection success' + str(self.client))
    def create_database(self,db_name):
        try:
            self.database=self.client[str(db_name)]
        except Exception as e:
            logging.exception('Error occured' + str(e))
            print('Error came' + str(e))
        else:
            logging.info('Database created' + str(self.database))
            print('Database created' + str(db_name))
    def create_collection(self,collection_name):
        try:
            self.collection=self.database[str(collection_name)]
        except Exception as e:
            print('Error came' + str(e))
            logging.exception('Error occured' + str(e))
        else:
            logging.info('Collection created'+ str(self.collection))
            print('Collection created' + str(collection_name))
    def insertion_data(self,f1):
        try:
            if type(f1)==list:
                self.collection.insert_many(f1)
            elif type(f1)==dict:
                self.collection.insert_one(f1)
        except Exception as e:
            print('Error came ' + str(e))
            logging.exception('Error came' + str(e))
        else:
            logging.info('Inserion success')
            print('Success')
    def update(self,pd,nd,update_all=True):
        try:
            if update_all==True:
                self.collection.update_many(pd,nd)
            elif update_all==False:
                self.collection.update_one(pd,nd)
        except Exception as e:
            print('Error came' + str(e))
            logging.exception('Error occured' + str(e))
        else:
            print('Update success')
            logging.info('Update occured')
    def delete(self,delete_data,delete_all=True):
        try:
            if delete_all==True:
                self.collection.delete_many(delete_data)
            elif delete_all==False:
                self.collection.delete_one(delete_data)
        except Exception as e:
            print('Error came' + str(e))
            logging.exception('Error has occured' + str(e))
        else:
            logging.info('Delete successfull')
            print('Delete success')
    def find(self,find_data):
        try:
            self.collection.find(find_data)
        except Exception as e:
            print('Error occured' + str(e))
            logging.exception('Error came' + str(e))
        else:
            logging.info('Searched one found')
            print('Searched data found')
            return self.collection.find(find_data)
    def filter(self,filter_data):
        try:
            self.collection.find(filter_data)
        except Exception as e:
            print('Error occured' + str(e))
            logging.exception('Error came' + str(e))
        else:
            logging.info('Searched one found')
            print('Searched data found')
            return self.collection.find(filter_data)
```
